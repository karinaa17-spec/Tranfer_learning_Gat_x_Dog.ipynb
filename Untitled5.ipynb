```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Caminho da imagem
caminho_imagem = '/content/cachorro.jpg'

# Verificar se o caminho do arquivo é válido
if not os.path.exists(caminho_imagem):
    print(f"O arquivo não foi encontrado no caminho: {caminho_imagem}")
else:
    # Carregar a imagem colorida
    imagem_colorida = cv2.imread(caminho_imagem)

    # Verificar se a imagem foi carregada corretamente
    if imagem_colorida is None:
        print(f"Erro ao carregar a imagem. Verifique o caminho: {caminho_imagem}")
    else:
        imagem_colorida = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)  # Converter para RGB

        # 1. Converter a imagem para níveis de cinza
        imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_RGB2GRAY)

        # 2. Binarizar a imagem (utilizando um limiar)
        limiar = 128
        _, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)

        # Exibir as imagens
        def mostrar_imagens(imagens, titulos):
            plt.figure(figsize=(10, 10))
            for i, img in enumerate(imagens):
                plt.subplot(1, len(imagens), i+1)
                plt.imshow(img, cmap='gray' if len(img.shape) == 2 else None)
                plt.title(titulos[i])
                plt.axis('off')
            plt.show()

        mostrar_imagens(
            [imagem_colorida, imagem_cinza, imagem_binarizada],
            ['Imagem Colorida', 'Imagem em Níveis de Cinza', 'Imagem Binarizada']
        )
```
